```python
import fastf1 as f1
import pandas as pd
```

```python
# Import weekend data from FastF1
YEAR = 2025
GP = 18

gp = f1.get_event(YEAR, GP)

if gp.EventFormat == 'conventional':
    fp1 = gp.get_session('Practice 1')
    fp1.load()
    fp2 = gp.get_session('Practice 2')
    fp2.load()
    fp3 = gp.get_session('Practice 3')
    fp3.load()
    ss = None
    sprint = None
    qualy = gp.get_session('Qualifying')
    qualy.load()
    race = gp.get_session('Race')
    race.load()
else:
    fp1 = gp.get_session('Practice 1')
    fp1.load()
    fp2 = None
    fp3 = None
    ss = gp.get_session('Sprint Shootout')
    ss.load()
    sprint = gp.get_session('Sprint')
    sprint.load()
    qualy = gp.get_session('Qualifying')
    qualy.load()
    race = gp.get_session('Race')
    race.load()
```

```python
# Convert session results to DataFrames
qualy_res = qualy.results
qualy_res.head()
```

```python
# Clean the results DataFrame as needed
qualy_res.drop(columns=['BroadcastName', 'DriverId', 'TeamColor', 'TeamId', 'FirstName', 'LastName', 'HeadshotUrl', 'CountryCode'], inplace=True)
```
